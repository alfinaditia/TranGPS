```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import mean_squared_error
from math import sqrt

# Melakukan Qcut, untuk memisahkan track.popularity
# dan Membaca data dan menetapkan kolom 'track.popularity' sebagai label
df = pd.read_csv('/content/datanew.csv')
df['track.popularity'] = pd.qcut(df['track.popularity'], q=2, labels=[0,1])
y = df['track.popularity']
print(df)

# Memilih fitur untuk digunakan dalam model
x = df[["speechiness","liveness","energy","danceability","tempo"]]

# Menkalibrasi fitur
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
print(x)
# Membagi data menjadi data latih dan data uji
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size=0.3, random_state=42)

# Membuat model SVM dan melatihnya dengan data latih
x_model = SVC()
x_model.fit(x_train, y_train)

# Mencetak skor akurasi model
x_acc = x_model.score(x_test, y_test)

# Memprediksi label dari data uji
y_pred = x_model.predict(x_test)

df_new = pd.DataFrame({'track.name': df.iloc[y_test.index]['track.name'],
                       'track.popularity': df.iloc[y_test.index]
                       ['track.popularity'], 'y_pred': y_pred})
df_new = df_new[df_new['y_pred'] == 1]

print("Akurasi menggunakan SVM adalah: ",x_acc)
print("")
print("Lagu yang diprediksi akan")
print("populer oleh algoritma SVM")
print("")
print(df_new)
print("")

mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error: ", mse)
```
